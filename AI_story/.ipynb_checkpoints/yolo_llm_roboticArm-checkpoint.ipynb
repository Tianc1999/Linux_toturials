## Import necessary library

```python
import cv2
import os 
import sys
import re
import time
import json
import base64
import requests
import numpy as np
import matplotlib.pyplot as plt

from openai import OpenAI
from yolov8.yolov8 import YOLOv8
from yolov8.utils import class_names

# library for robotic arm control 
from AI.tutorials.lib.Arm7Bot import Arm7Bot
arm = Arm7Bot("COM3")
```

## Setup yolov8 and video stream

```python
video_cap = cv2.VideoCapture(0)
video_cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
video_cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

# Create an inference session using the Vitis AI execution provider
input_model_file = r"onnx_utils\DetectionModel_int.onnx"
config_file_path = r"onnx_utils\vaip_config.json"
model = YOLOv8(input_model_file, config_file_path, 0.25) # probability threshold: 0.25. Notes: high probability threshold will casue the model failing in detecting object.

# capture one image
success, frame = video_cap.read()
frame = frame[:, 280:-280, ::-1]

print(frame.shape)
plt.imshow(frame)
plt.imsave('temp\out.jpg', frame)

ret = model.detect_objects(frame[:, :, ::-1])
detected_img = model.draw_detections(frame)
plt.imshow(detected_img)
```

```python
def parse_ret(ret, origin_img):
    parsed_data = []
    parsed_img = {}
    for i in range(len(ret[0])):
        x1, y1, x2, y2 = ret[0][i].astype(np.int32)
        object_dict = {
            "Index": "ThisIsSearchIndex_"+str(i),
            "coordination": [str((x1+x2)//2), str((y1+y2)//2)],
            "class": class_names[ret[2][i]],
        }
        parsed_img["ThisIsSearchIndex_"+str(i)] = origin_img[y1:y2, x1:x2, :]
        parsed_data.append(object_dict)
    return parsed_data, parsed_img


parsed_result, parsed_img = parse_ret(ret, frame[:, :, ::1])
print(parsed_result)
```

## ENTER YOUR QUERY HERE

```python
query = "which gift is better for my little daughter?"
```

### This our enhanced prompt here. 
This prompt is aimed at link the objects in image with the detection results from yolov8 so that the model can ground target object precisely.

```python
objects = json.dumps(parsed_result, indent=4)
prompt_1 = f"""
You are a helpful AI agent with camera. 
You will be provided with a list of <objects> in one image. 

###### INPUT DESCRIPTION ######
<objects>: Detected objects in the image. This is formatting in json.
""""""
{
    {
        'Index': 'ThisIsSearchIndex_1',
        'coordination': [x1, y1],
        'class': '...',
    },
    {
        'Index': 'ThisIsSearchIndex_2',
        'coordination': [x2, y2],
        'class': '...',
    }, 
}
""""""

REMEMBER: the coordination stands for the position of the object in the image.

###### TASK INSTRUCTION ######
You need to match the user's query and return one of the <objects> and its Index.

###### OUTPUT INSTRUCTION ######
Remember to explain the reason step-by-step and return the object Index at the end.

For example:

The reason is <your reason here> 
The answer is <your answer here> and its index is <ThisIsSearchIndex_Index>.

###### The following is the real input ######
Remember: You must return only a answer and its Index. Your must answer based on the objects provided and the image.

"""

prompt_2 = f"""
###### INPUT ######
<objects>: {objects} 

###### USER QUERY ###### 
User query: {query}. Let's do it step-by-step. Please return its Index in the end. You must return one answer.

###### OUTPUT ######
Answer: 
"""
prompt = prompt_1 + prompt_2
```

## Call LLM to reason
The LLM api is called in format of openai. And the model we use here is the "llava-7b".

```python
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
  
# Path to your image
image_path = "temp\out.jpg"

# Getting the base64 string
base64_image = encode_image(image_path)

client = OpenAI(base_url="http://127.0.0.1:1234/v1")
messages = []
first_turn = {
            "role": "user",
            "content": [
                {
                "type": "text",
                "text": f"{prompt}",
                },
                {
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/jpeg;base64,{base64_image}"
                }
                }
            ]
        }
messages.append(first_turn)
payload = {
    "model": "llava-v1.5-7b-q4_k",
    "messages": f"{json.dumps(messages, indent=4)}",
}
response = requests.post("http://127.0.0.1:1234/v1/chat/completions", json=payload)
print(response)
response.json()
```

# THIS IS THE RESULT
We use re to get the answer. Note: this is a very simple method, we recommend you to tune prompt to ask LLM return answer in json or other method to extract the final answer.

```python
pattern_string = r"ThisIsSearchIndex_\d+"
matched_string = re.findall(pattern_string, response.text.replace('\\', ''))[0]

ans = None 
for elem in parsed_result:
    if elem['Index'] == matched_string:
        ans = elem 
        break

print(elem)
plt.imshow(parsed_img[elem['Index']])
bias_y = 20
bias_x = 30
coords = [int(elem['coordination'][0])+bias_x, int(elem['coordination'][1])+bias_y, 1]
```

## Setup robotic arm here

```python
if arm.ser.isOpen() == False:
    arm.ser.open()
    
arm.setStatus(1)
arm.setSpeed(20)
```

```python
def grab_item(pos6, vec56, vec67):
    arm.setSpeed(5)
    arm.setAngle(6, 90)
    time.sleep(2)
    origin_height = pos6[-1]
    pos6[-1] = 20

    arm.setIK7(pos6, vec56, vec67)
    time.sleep(2)
    arm.setAngle(6, 15)
    time.sleep(4)
    
    pos6[-1] = origin_height
    arm.setIK7(pos6, vec56, vec67)
    time.sleep(2)
    return 

def release_item(pos6, vec56, vec67):
    origin_height = pos6[-1]
    pos6[-1] = 20

    arm.setIK7(pos6, vec56, vec67)
    time.sleep(2)
    arm.setAngle(6, 90)
    time.sleep(4)
    
    pos6[-1] = origin_height
    arm.setIK7(pos6, vec56, vec67)
    time.sleep(2)
    return
```

```python
# calculate transform affine here
#
objpoints = np.array([
    [180, 160],
    [200, 20],
    [280, 110],
], dtype=np.float32)

imgpoints = np.array([
    [200, 289],
    [597, 372],
    [412, 586]
], dtype=np.float32)

out = cv2.getAffineTransform(imgpoints, objpoints,)
tgt_point = np.dot(out, np.array(coords)).astype(np.int32)

if arm.ser.isOpen() == False:
    arm.ser.open()

time.sleep(2)

arm.setIK7([tgt_point[0], tgt_point[1], 150], [0, 0, -1], [0, -1, 0])
time.sleep(5)
arm.setIK7([tgt_point[0], tgt_point[1], 100], [0, 0, -1], [0, -1, 0])

grab_item([tgt_point[0], tgt_point[1], 100], [0, 0, -1], [0, -1, 0])
release_item([tgt_point[0], tgt_point[1], 100], [0, 0, -1], [0, -1, 0])

if arm.ser.isOpen():
    arm.ser.close()
```
